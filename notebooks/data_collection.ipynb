```python
languages = {"hindi" : "hi",
             "gujarati" : "gu",
             "punjabi" : "pa",
             "marathi": "mr",
             "tamil": "ta",
             "telugu": "te",
             "malayalam": "ml",
             "kannada": "kn"
            }
```

```python
for lang, short in languages.items():
    
    file_path = os.path.join(PATH, f"{lang}.csv")
    
    print(f"Start downloading {lang.capitalize()} data...")
    full_dataset = load_dataset("ai4bharat/IndicHeadlineGeneration", short, trust_remote_code=True)
    print(f"Download complete...")
    
    short_dataset = full_dataset["train"].to_pandas()["input"].iloc[:500]
    
    print(f"{lang.capitalize()} data stored as csv...")
    short_dataset.to_csv(file_path, index=False)
```

```python
all_data = pd.DataFrame()

for lang, short in languages.items():
    
    df = pd.read_csv(PATH + f"/{lang}.csv")
    df["label"] = [lang.capitalize()] * len(df)
    
    all_data = pd.concat([all_data, df], axis=0)
```
